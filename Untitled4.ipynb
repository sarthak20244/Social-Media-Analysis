## Importing Data Set (Student Social Media Addiction) from Kaggle and using csv file

```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv(r"C:\Users\lenovo\Desktop\Students Social Media Addiction.csv")
df
```

# Dropping Unwanted Columns 

```python
df.drop(['Relationship_Status','Conflicts_Over_Social_Media'],axis=1,inplace=True)
df
```

## Renaming column header for better visbility 

```python
df.rename(columns={'Student_ID': 'Serial_No','Avg_Daily_Usage_Hours': 'Average_Usage','Most_Used_Platform': 'Social_Media','Affects_Academic_Performance': 'Academic_Performance','Sleep_Hours_Per_Night': 'Sleep','Addicted_Score': 'Addiction'},inplace=True)
df
```

```python
df.head(50)
```

## Getting Distinct Values from different Columns

```python
df.nunique()
```

## Checking for Null Values

```python
df.isnull().sum()
```

## Sorting Columns Age,Average_Usage,Sleep in Ascending order

```python
df.sort_values(by=['Age','Average_Usage','Sleep'])
```

## Correlation Between Gender and Average_Usage of Social Media

```python
gender_series = df['Gender']
avg_usage = df['Average_Usage'].to_numpy()

gender_numeric = gender_series.map({'Male': 0, 'Female': 1}).to_numpy()

correlation = np.corrcoef(gender_numeric, avg_usage)[0, 1]
correlation
```

## Bar plot of mean usage by gender

```python
sns.barplot(x='Gender', y='Average_Usage', data=df)
plt.title('Average Daily Usage Hours by Gender')
plt.ylabel('Avg Daily Usage (Hours)')
plt.show()
```

## Heatmap

```python
pivot_table = df.groupby('Gender')['Average_Usage'].agg(['mean', 'min', 'max'])

print(pivot_table)

# Plot as heatmap
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu')

plt.title('Heatmap: Avg Daily Usage by Gender')
plt.xlabel('Metric')
plt.ylabel('Gender')
plt.show()
```

## Average Usage Vs Academic,Sleep,Mental Health,Addiction

```python
df['Academic_Performance'] = df['Academic_Performance'].map({'Yes': 1, 'No': 0})

cols = ['Average_Usage', 'Academic_Performance', 'Sleep', 'Mental_Health_Score', 'Addiction']

corr_matrix = df[cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix: Average Usage vs Academic, Sleep, Mental Health, Addiction')
plt.show()
```

# Deeper Analysis

## Demographic Analysis-Age Distribution 

```python
sns.histplot(df['Age'], bins=10, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

counts, bins = np.histogram(df['Age'], bins=10)
for i in range(len(counts)):
    plt.text(bins[i] + (bins[1]-bins[0])/2, counts[i], str(counts[i]), ha='center')

plt.show()
```

 ## Demographic Analysis- Gender Ratio

```python
gender_counts = df['Gender'].value_counts()

gender_counts.plot(
    kind='pie',
    autopct=lambda p: '{:.1f}%\n({:.0f})'.format(p, p * gender_counts.sum() / 100),
    colors=['skyblue', 'pink']
)
plt.title('Gender Ratio (with counts)')
plt.ylabel('')
plt.show()
```

## Age and its relation with Average Usage

```python
bins = [10, 15, 20, 25, 30]
labels = ['10-15', '16-20', '21-25', '26-30']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

plot = sns.barplot(x='Age_Group', y='Average_Usage', data=df, errorbar=None)
plt.title('Average Daily Usage by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Usage (Hours)')

for p in plot.patches:
    plot.annotate(f"{p.get_height():.2f}",
                  (p.get_x() + p.get_width()/2, p.get_height()),
                  ha='center', va='bottom')

plt.show()
```

## Gender vs Academic Performance & Usage

```python
sns.boxplot(x='Gender', y='Average_Usage', hue='Academic_Performance', data=df)
plt.title('Usage by Gender and Academic Performance')
plt.show()
```

## Most Used Platform

```python
plot = sns.countplot(y='Social_Media', data=df, order=df['Social_Media'].value_counts().index)
plt.title('Most Used Social Media Platforms')
plt.xlabel('Count')
plt.ylabel('Platform')

for p in plot.patches:
    width = p.get_width()
    plt.text(width + 0.5, p.get_y() + p.get_height()/2, int(width), va='center')

plt.show()
```

## Addiction Score by Platform

```python
import seaborn as sns
import matplotlib.pyplot as plt

platform_means = df.groupby('Social_Media')['Addiction'].mean().reset_index()

sns.barplot(x='Social_Media', y='Addiction', data=platform_means)

for index, row in platform_means.iterrows():
    plt.text(index, row['Addiction'] + 0.05, f"{row['Addiction']:.2f}", ha='center')

plt.xticks(rotation=30)

plt.title('Mean Addiction Score by Platform')
plt.xlabel('Social Media Platform')
plt.ylabel('Mean Addiction Score')
plt.show()
```
